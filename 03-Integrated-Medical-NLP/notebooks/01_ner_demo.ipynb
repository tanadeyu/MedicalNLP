# 01: NER単体デモ

BioBERTでテキストから疾患・薬剤を抽出する簡単なデモ。

## 1. ライブラリインポート

```python
import torch
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline

print("インポート完了")
```

## 2. NERモデル読み込み

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model_path = "../models/biobert-ner-bc5cdr"

# パイプライン作成
# aggregation_strategy="first" は WordPiece サブワードを正しく結合
ner_pipeline = pipeline(
    "token-classification",
    model=model_path,
    tokenizer=model_path,
    aggregation_strategy="first",
    device=0 if device == "cuda" else -1
)

print(f"NERモデル読み込み完了: {device}")
print(f"aggregation_strategy: first (WordPiece サブワード結合用)")
```

## 3. エンティティ抽出

```python
# サンプルテキスト
text = "Patient has diabetes and hypertension. He takes insulin and metformin."

# NER実行
entities = ner_pipeline(text)

# 結果表示
print(f"テキスト: {text}\n")
print("抽出されたエンティティ:")
for ent in entities:
    print(f"  - {ent['entity_group']}: {ent['word']}")
```
